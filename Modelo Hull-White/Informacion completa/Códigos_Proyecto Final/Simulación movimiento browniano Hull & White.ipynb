A continuación se presentan las simulaciones de trayectorias de movimientos brownianos:

```python
#Importamos las librerías necesarias
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la simulación
T = 1.0        # Tiempo total
N = 500        # Número de pasos
dt = T / N     # Tamaño del paso de tiempo
t = np.linspace(0, T, N+1)  # Vector de tiempos

# Número de trayectorias a simular
num_trayectorias = 5

# Inicializar matriz para guardar trayectorias
trayectorias = np.zeros((num_trayectorias, N+1))

# Simulación de las trayectorias
for j in range(num_trayectorias):
    dB = np.random.normal(0.0, np.sqrt(dt), N)   # Incrementos normales ~ N(0, dt)
    B = np.zeros(N+1)
    B[1:] = np.cumsum(dB)   # Construir trayectoria a partir de los incrementos
    trayectorias[j] = B

# Graficar las trayectorias
plt.figure(figsize=(10, 6))
for j in range(num_trayectorias):
    plt.plot(t, trayectorias[j], label=f'Trayectoria {j+1}')
plt.title('Simulación de trayectorias del movimiento Browniano')
plt.xlabel('Tiempo')
plt.ylabel('B(t)')
plt.grid(True)
plt.legend()

# Guardar la figura como imagen PNG
plt.savefig("trayectorias_browniano.png", dpi=300)

# Mostrar la figura
plt.show()
```
